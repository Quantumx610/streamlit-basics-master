```python
df = pd.read_csv('startup_funding.csv')
```

```python
df.head()
```

```python
l = sorted(df['Startup Name'].unique().tolist())
```

```python
df.info()
```

```python
df['Investors Name']  = df['Investors Name'].fillna('Undisclosed')
```

```python
df.head()
```

```python
sorted(df['Investors Name'].unique().tolist())
```

```python
df.drop(columns = ['Remarks'], inplace = True)
```

```python
df.set_index('Sr No',inplace= True)
```

```python
df.rename(columns= {
    'Date dd/mm/yyyy' : 'Date',
    'Startup Name' : 'Startup',
    'Industry Vertical' : 'Verticle',
    'City  Location' : 'City',
    'Investors Name' : 'Investors',
    'InvestmentnType' : 'Round',
    'Amount in USD' : 'Amount',
}, inplace = True)
```

```python
df.head()
```

```python
df.info()
#amount is string
```

```python
#converting the undisclosed investement to zero
df['Amount'] = df['Amount'].fillna('0')
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].str.replace(',', '')
df['Amount'] = df['Amount'].str.replace('Undisclosed', '0')
df['Amount'] = df['Amount'].str.replace('Unknown', '0')
df['Amount'] = df['Amount'].str.replace('undisclosed', '0')
df['Amount'] = df['Amount'].str.replace('unknown', '0')
```

```python
df['Amount']
```

```python
df = df[df['Amount'].str.isdigit()]
```

```python
df['Amount'] = df['Amount'].astype('float')
```

```python
df['Amount']
```

```python
df.head()
```

```python
def to_inr(dollar):
    inr = dollar* 82.5
    return inr/10000000
```

```python
df['Amount']= df['Amount'].apply(to_inr)
```

```python
df['Amount']
```

```python
df['Date'] = df['Date'].str.replace('05/072018','05/07/2018')
```

```python
#date time object
df['Date']= pd.to_datetime(df['Date'],format='%d/%m/%Y',errors='coerce')
```

```python
df.info()
```

```python
df['Date'].dt.day
```

```python
df = df.dropna(subset= ['Date','Startup','Verticle','City','Round', 'Investors','Amount'])
```

```python
df.info()
```

```python
#exporting a data

df.to_csv('Startup_cleaned.csv', index= False)
```
